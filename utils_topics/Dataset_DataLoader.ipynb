```python
file_path = "/content/drive/MyDrive/dataset/wine.csv"
# file_path = "/content/drive/MyDrive/dataset/dataTest.csv"
```

```python
class WineDataset(Dataset):
    def __init__(self):
        # data loading
        xy = np.genfromtxt(file_path, delimiter=",", dtype=np.float32, skip_header=1)
        self.x = torch.from_numpy(xy[:, 1:])
        self.y = torch.from_numpy(xy[:, [0]])
        self.n_samples = xy.shape[0]
        # print(xy)
        # print(self.n_samples)
        # print(self.x.shape)
        # print(self.x)
        # print(self.y.shape)
        # print(self.y)
    #Obligatorio implementarla para hacer un llamado a => dataset[index]
    def __getitem__(self, index):
        return self.x[index], self.y[index]

    # Implementar obligatoriamente para el uso de len(dataset)
    def __len__(self):
        # return len(dataset)
        return self.n_samples
```

```python
winedataset = WineDataset()
print(len(winedataset))
```

```python
print(winedataset[0])
firstData = winedataset[0]
features, labels = firstData
print(features, labels)
```

```python
dataloader = DataLoader(dataset=winedataset, batch_size=5, shuffle=True, num_workers=2)
```
