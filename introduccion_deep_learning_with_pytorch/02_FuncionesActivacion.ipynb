De la imagen anterior hay que recalcar que la funci√≥n:
\begin{align*}
    f(x) = e^x
\end{align*}
tiene el siguiente comportamiento:
\begin{align*}
    \lim_{x\to\infty} f(x) &= \infty \\
    \lim_{x\to-\infty} f(x) &= 0
\end{align*}

Ademas, que la funci√≥n:
\begin{align*}
    g(x) = e^{-x}
\end{align*}
tiene el siguiente comportamiento:
\begin{align*}
    \lim_{x\to\infty} g(x)&= 0 \\
    \lim_{x\to-\infty} g(x)&= \infty
\end{align*}


```python
import plotly.graph_objects as go
import numpy as np

def my_sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Define the range of x-axis values
x = np.linspace(-10, 10, 400)  # 400 points for smooth curves

# Calculate sine and cosine values
#y_sin = np.sin(x)
#y_cos = np.cos(x)
y_sigmoidea = my_sigmoid(x)

# Create the plot
fig = go.Figure()

# Plot the sine function (red line)
# fig.add_trace(go.Scatter(x=x, y=y_sin, mode='lines', name='Sine', line=dict(color='red')))

fig.add_trace(go.Scatter(x=x, y=y_sigmoidea, mode='lines', name='Singmoidea', line=dict(color='red')))

# Plot the cosine function (blue line)
#fig.add_trace(go.Scatter(x=x, y=y_cos, mode='lines', name='Cosine', line=dict(color='blue')))

# Add labels and title
fig.update_layout(
    title='Sigmoidea Function',
    xaxis_title='X',
    yaxis_title='Y'
)

# Show the plot
fig.show() # to colab
name = "sigmoidea.png"
saveImage_toShowGitHub(name, fig)
load_image_path = "images/" + name
Load_show_image(load_image_path) #to github
```

De la funcion sigmoidea:
\begin{align*}
    h(x) = \frac{1}{1+e^{-x}}
\end{align*}

La funcion anterior tiene el siguiente comportamiento cuando $x\to\infty$:

\\
\begin{align*}
    \lim_{x\to\infty} h(x) &= 1
\end{align*}
La explicaci√≥n es la siguiente:
*   cuando $x\to\infty$, $e^{-x}$ se hace 0, por lo tanto $h(x) = \frac{1}{1+0} = 1$.

\\
Luego, la funcion anterior tiene el siguiente comportamiento cuando $x\to -\infty$:
\begin{align*}
    \lim_{x\to-\infty} h(x)&= 0
\end{align*}
La explicaci√≥n es la siguiente:
*   cuando $x\to -\infty$, $e^{-x}$ se hace $\infty$, de aqui el denominador de $h(x)$ se hace $\infty$, por lo tanto $h(x) = \frac{1}{\infty} = 0$.

## **Conociendo Softmax**

##### **Softmax function**
La funci√≥n softmax œÉ: $\mathbb{R}^{k}$ ‚ü∂ $[0,1]^{k}$, para todo $a = (a_1, a_2, ... , a_k)^T \in \mathbb{R}^{k}$.

\begin{align*}
    œÉ(a) &= \begin{pmatrix}
        œÉ(a)_1\\
        œÉ(a)_2\\
        œÉ(a)_3\\
        \vdots \\
        œÉ(a)_k\\
            \end{pmatrix}
\end{align*}

Con,
\begin{align*}
    œÉ(a)_k &= \frac{e^{a_k}}{\sum_{j=1}^{k} e^{a_j}}
 \end{align*}

```python
# prompt: implementacion de softmax function
import numpy as np
def softmax(x):
  """
  This function computes the softmax of a vector x.
  """
  exp_x = np.exp(x)
  sum_exp_x = np.sum(exp_x)
  y = exp_x / sum_exp_x
  return y

# Define the input vector
x = np.array([4.3, 6.1, 2.3])

# Compute the softmax of the vector
y = softmax(x)

# Print the softmax values
print(y)
import math

print(math.exp(0))
print(math.exp(2))
print(2.7182*2.7182)
```

```python
# prompt: implementacion de la funcion softmax con tensor de pytorch

import torch
import torch.nn as nn

def softmax_torch(x):
  """
  This function computes the softmax of a vector x using PyTorch.
  """
  softmax = nn.Softmax(dim=-1)
  y = softmax(x)
  return y

# Define the input vector
x = torch.tensor([4.3, 6.1, 2.3])

# Compute the softmax of the vector
y = softmax_torch(x)

# Print the softmax values
print(y)
print(y.dtype)
```
