De la imagen anterior hay que recalcar que la funci√≥n:
\begin{align*}
    f(x) = e^x
\end{align*}
tiene el siguiente comportamiento:
\begin{align*}
    \lim_{x\to\infty} f(x) &= \infty \\
    \lim_{x\to-\infty} f(x) &= 0
\end{align*}

Ademas, que la funci√≥n:
\begin{align*}
    g(x) = e^{-x}
\end{align*}
tiene el siguiente comportamiento:
\begin{align*}
    \lim_{x\to\infty} g(x)&= 0 \\
    \lim_{x\to-\infty} g(x)&= \infty
\end{align*}


```python
import plotly.graph_objects as go
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Define the range of x-axis values
x = np.linspace(-10, 10, 400)  # 400 points for smooth curves

# Calculate sine and cosine values
#y_sin = np.sin(x)
#y_cos = np.cos(x)
y_sigmoidea = sigmoid(x)

# Create the plot
fig = go.Figure()

# Plot the sine function (red line)
# fig.add_trace(go.Scatter(x=x, y=y_sin, mode='lines', name='Sine', line=dict(color='red')))

fig.add_trace(go.Scatter(x=x, y=y_sigmoidea, mode='lines', name='Singmoidea', line=dict(color='red')))

# Plot the cosine function (blue line)
#fig.add_trace(go.Scatter(x=x, y=y_cos, mode='lines', name='Cosine', line=dict(color='blue')))

# Add labels and title
fig.update_layout(
    title='Sigmoidea Function',
    xaxis_title='X',
    yaxis_title='Y'
)

# Show the plot
fig.show()
```

De la funcion sigmoidea:
\begin{align*}
    h(x) = \frac{1}{1+e^{-x}}
\end{align*}

La funcion anterior tiene el siguiente comportamiento cuando $x\to\infty$:

\\
\begin{align*}
    \lim_{x\to\infty} h(x) &= 1
\end{align*}
La explicaci√≥n es la siguiente:
*   cuando $x\to\infty$, $e^{-x}$ se hace 0, por lo tanto $h(x) = \frac{1}{1+0} = 1$.

\\
Luego, la funcion anterior tiene el siguiente comportamiento cuando $x\to -\infty$:
\begin{align*}
    \lim_{x\to-\infty} h(x)&= 0
\end{align*}
La explicaci√≥n es la siguiente:
*   cuando $x\to -\infty$, $e^{-x}$ se hace $\infty$, de aqui el denominador de $h(x)$ se hace $\infty$, por lo tanto $h(x) = \frac{1}{\infty} = 0$.
